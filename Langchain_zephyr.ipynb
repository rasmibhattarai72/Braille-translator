```python
from langchain.document_loaders.csv_loader import CSVLoader
```

```python
loader = CSVLoader(file_path='/content/faq-him.csv')
data = loader.load()
```

```python
print(data)
```

```python
import os
import getpass
```

```python
from langchain.vectorstores import FAISS
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
```

```python
!pip install faiss-cpu
```

```python
!pip install sentence-transformers
```

```python
db = FAISS.from_documents(data, HuggingFaceEmbeddings(model_name = "HuggingFaceH4/zephyr-7b-alpha"))
```

```python
db.save_local("/content/vecStore.csv")
```

```python
db_embed = db.load_local("/content/vecStore.csv", HuggingFaceEmbeddings(model_name = "HuggingFaceH4/zephyr-7b-alpha"))
```

```python
query = "How many days are needed to settle down the claim?"
docs = db_embed.similarity_search(query)
print(docs)
```
