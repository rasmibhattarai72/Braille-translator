```python
CAPITAL = chr(10272)  # ⠠
NUMBER = chr(10300)  # ⠼
UNRECOGNIZED = '?'
open_quotes = True


def extract_words_A(string):
    words = string.split(" ")
    result = []
    for word in words:
        temp = word.split("\n")
        for item in temp:
            result.append(item)
    return result


def is_braille_A(char):
    if len(char) > 1:
        return False
    return char in letters_B \
        or char in numbers_B \
        or char in punctuation_B \
        or char in contractions_B \
        or char == CAPITAL \
        or char == NUMBER


def trim_A(word):
    while len(word) != 0 and not word[0].isalnum():
        word = word[1:]
    while len(word) != 0 and not word[-1].isalnum():
        word = word[:-1]
    return word


def numbers_handler_A(word):
    if word == "":
        return word
    result = word[0]
    if word[0].isdigit():
        result = NUMBER + numbers_A.get(word[0])
    for i in range(1, len(word)):
        if word[i].isdigit() and word[i-1].isdigit():
            result += numbers_A.get(word[i])
        elif word[i].isdigit():
            result += NUMBER + numbers_A.get(word[i])
        else:
            result += word[i]
    return result


def capital_letters_handler_A(word):
    if word == "":
        return word
    result = ""
    for char in word:
        if char.isupper():
            result += CAPITAL + char.lower()
        else:
            result += char.lower()
    return result


def find_utf_code_A(char):
    if len(char) != 1:
        return -1
    for i in range(0, 55000):
        if char == chr(i):
            return i


def char_to_braille_A(char):
    if is_braille_A(char):
        return char
    elif char == "\n":
        return "\n"
    elif char == "\"":
        global open_quotes
        if open_quotes:
            open_quotes = not open_quotes
            return punctuation_A.get("“")
        else:
            open_quotes = not open_quotes
            return punctuation_A.get("”")
    elif char in letters_A and char.isupper():
        return CAPITAL + letters_A.get(char)
    elif char in letters_A:
        return letters_A.get(char)
    elif char in punctuation_A:
        return punctuation_A.get(char)
    else:
        print("Unrecognized Symbol:", char, "with UTF code:", find_utf_code(char))
        return UNRECOGNIZED


def word_to_braille_A(word):
    if word in contractions_A:
        return contractions_A.get(word)
    else:
        result = ""
        for char in word:
            result += char_to_braille_A(char)
        return result


def build_braille_word_A(trimmed_word, shavings, index, braille):
    if shavings == "":
        braille += word_to_braille_A(trimmed_word)
    else:
        for i in range(0, len(shavings)):
            if i == index and trimmed_word != "":
                braille += word_to_braille_A(trimmed_word)
            braille += word_to_braille_A(shavings[i])
        if index == len(shavings): 
            braille += word_to_braille_A(trimmed_word)
    return braille


def translate_A(string):
    braille = ""
    words = extract_words_A(string)
    for word in words:
        word = numbers_handler_A(word)
        word = capital_letters_handler_A(word)
        trimmed_word = trim_A(word)  
        untrimmed_word = word
        index = untrimmed_word.find(trimmed_word)
        shavings = untrimmed_word.replace(trimmed_word, "")
        braille = build_braille_word_A(trimmed_word, shavings, index, braille) + " "
    return braille[:-1] 
```

```python
def extract_words_B(string):
    words = string.split(" ")
    result = []
    for word in words:
        temp = word.split("\n")
        for item in temp:
            result.append(item)
    return result


def trim_B(word):
    while len(word) != 0 and word[0] not in letters_B \
            and word[0] not in contractions_B:
        word = word[1:]
    while len(word) != 0 and word[-1] not in letters_B \
            and word[-1] not in contractions_B:
        word = word[:-1]
    return word


def numbers_handler_B(word):
    if word == "":
        return word
    result = ""
    is_number = False
    for i in range(0, len(word)):
        if word[i] == NUMBER:
            is_number = True
            continue
        if is_number and word[i] in numbers_B:
            result += numbers_B.get(word[i])
        else:
            result += word[i]
    return result


def capital_letters_handler_B(word):
    if word == "":
        return word
    result = ""
    for i in range(0, len(word)):
        if i-1 >= 0 and word[i-1] == CAPITAL and word[i] in letters_A:
            result += word[i].upper()
        elif word[i] != CAPITAL:
            result += word[i]
    return result


def word_to_alpha_B(word):
    if word in contractions_B:
        return contractions_B.get(word)
    else:
        result = ""
        for i in range(0, len(word)):
            if word[i] in letters_B:
                result += letters_B.get(word[i])
            elif word[i] in punctuation_B:
                result += punctuation_B.get(word[i])
        return result


def fix_exceptions_B(string):
    result = ""
    for i in range(0, len(string)):
        if i-1 >= 0 and string[i] == "“" and string[i-1] in letters_A:
            result += "?"
        else:
            result += string[i]

    open_bracket = False
    while result.find("()") != -1:
        if open_bracket:
            result = result.replace("()", ")", 1)
        else:
            result = result.replace("()", "(", 1)
        open_bracket = not open_bracket
    return result


def build_alpha_word_B(trimmed_word, shavings, index, alpha):
    if shavings == "":
        alpha += word_to_alpha_B(trimmed_word)
    else:
        for i in range(0, len(shavings)):
            if i == index and trimmed_word != "":
                alpha += word_to_alpha_B(trimmed_word)
            if shavings[i] in punctuation_B:
                alpha += punctuation_B.get(shavings[i])
            else:
                alpha += shavings[i]
        if index == len(shavings): 
            alpha += word_to_alpha_B(trimmed_word)
    return alpha


def translate_B(string):
    alpha = ""
    words = extract_words_B(string)
    for word in words:
        word = numbers_handler_B(word)
        trimmed_word = trim_B(word) 
        untrimmed_word = word
        index = untrimmed_word.find(trimmed_word)
        shavings = untrimmed_word.replace(trimmed_word, "")
        alpha = build_alpha_word_B(trimmed_word, shavings, index, alpha) + " "
    alpha = capital_letters_handler_B(alpha[:-1])
    return fix_exceptions_B(alpha)
```

```python
def uppercase_alphabet_utf_codes():
    print("UTF Codes for Capital Letters:")
    for i in range(65, 91):
        print(chr(i), i, end='    ')
        if i % 8 == 0:
            print()
    print()


def lowercase_alphabet_utf_codes():
    print("UTF Codes for Letters:")
    for i in range(97, 123):
        print("%c %3d" % (chr(i), i), end='   ')
        if i % 8 == 0:
            print()
    print()


def braille_utf_codes():
    print("UTF Codes for Braille Symbols:")
    for i in range(10240, 10304):
        print(chr(i), i, end='  ')
        if (i+1) % 8 == 0:
            print()


def alphabet():
    print("Letters:")
    for i in range(97, 123):
        print(chr(i), letters_A.get(chr(i)), end='   ')
        if i % 8 == 0:
            print()
    print()


def contractions():
    print("Contraction:")
    count = 0
    word_list = []
    for word in contractions_A:
        word_list.append(word)
    word_list.sort()
    for word in word_list:
        formatted = '{:<10}'.format(word)
        print("%c %10s" % (contractions_A.get(word), formatted), end="")
        count += 1
        if count % 5 == 0:
            print()
    print()


def numbers():
    print("Numbers (must proceed after the ⠼ escape code):")
    count = 0
    num_list = []  # Punctuation list.
    for num in numbers_A:
        num_list.append(num)
    num_list.sort()
    for num in num_list:
        print(num, numbers_A.get(num), end="        ")
        count += 1
        if count % 5 == 0:
            print()


def punctuation():
    print("Punctuation:")
    count = 0
    pun_list = []  # Punctuation list.
    for pun in punctuation_A:
        pun_list.append(pun)
    pun_list.sort()
    for pun in pun_list:
        print(pun, punctuation_A.get(pun), end="        ")
        count += 1
        if count % 5 == 0:
            print()
    print()


def all_utf_codes():
    lowercase_alphabet_utf_codes()
    print()
    uppercase_alphabet_utf_codes()
    print()
    braille_utf_codes()


def all_braille():
    alphabet()
    print()
    contractions()
    print()
    punctuation()
    print()
    numbers()
    print()
```
